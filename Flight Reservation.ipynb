3. Data Cleaning

```python
# Drop rows with missing values in specific columns
df.dropna(subset=['ActualElapsedTime', 'AirTime', 'ArrDelay', 'ArrTime', 'TaxiIn'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# Create a date column from Year, Month, and DayofMonth
df['Date'] = pd.to_datetime(df[['Year', 'Month', 'DayofMonth']].rename(columns={'DayofMonth': 'day', 'Month': 'month', 'Year': 'year'}))
df.info()
```

# Dim Airport

```python
# Keep Origin and Dest as they are and remove duplicates
dim_airport = df[['Origin', 'Dest']].drop_duplicates().reset_index(drop=True)

# Create surrogate key by combining Origin and Dest
dim_airport['Airport_ID'] = (dim_airport['Origin'] + '-' + dim_airport['Dest'])

# Save to CSV
dim_airport.to_csv('dim_airport.csv', index=False)

print(dim_airport.head())
print(dim_airport.shape)
```

# Dim Airline

```python
# Keep the 4 columns and remove duplicates
dim_airline = df[['UniqueCarrier', 'FlightNum', 'TailNum']].drop_duplicates().reset_index(drop=True)

# Cast to string and create surrogate key by combining the 3 columns
dim_airline['Airline_ID'] = (dim_airline['UniqueCarrier'].astype(str) + '-' + 
                              dim_airline['FlightNum'].astype(str) + '-' + 
                              dim_airline['TailNum'].astype(str))

# Save to CSV
dim_airline.to_csv('dim_airline.csv', index=False)

print(dim_airline.head())
print(dim_airline.shape)
```

# Dim Date

```python
# Keep the 4 columns and remove duplicates
dim_date = df[['Year', 'Month', 'DayofMonth', 'DayOfWeek']].drop_duplicates().reset_index(drop=True)

# Create Date column
dim_date['Date'] = pd.to_datetime(dict(year=dim_date['Year'], 
                                        month=dim_date['Month'], 
                                        day=dim_date['DayofMonth']))

# Create surrogate key by combining the 3 columns
dim_date['Date_ID'] = (dim_date['Year'].astype(str) + '-' + 
                        dim_date['Month'].astype(str) + '-' + 
                        dim_date['DayofMonth'].astype(str))

# Save to CSV
dim_date.to_csv('dim_date.csv', index=False)

print(dim_date.head())
print(dim_date.shape)
```

# Fact Flights

```python
fact_flights = df.drop(columns=['Unnamed: 0',
                                 'Year', 'Month', 'DayofMonth', 'DayOfWeek',
                                 'UniqueCarrier', 'FlightNum', 'TailNum',
                                 'Origin', 'Dest']).copy()

fact_flights['Date_ID'] = (df['Year'].astype(str) + '-' + 
                            df['Month'].astype(str) + '-' + 
                            df['DayofMonth'].astype(str))

fact_flights['Airport_ID'] = (df['Origin'].astype(str) + '-' + 
                               df['Dest'].astype(str))

fact_flights['Airline_ID'] = (df['UniqueCarrier'].astype(str) + '-' + 
                               df['FlightNum'].astype(str) + '-' + 
                               df['TailNum'].astype(str))

fact_flights.fillna(0, inplace=True)

# Optimize Data Types
fact_flights['DepTime'] = fact_flights['DepTime'].astype('int16')
fact_flights['ArrTime'] = fact_flights['ArrTime'].astype('int16')
fact_flights['CRSDepTime'] = fact_flights['CRSDepTime'].astype('int16')
fact_flights['CRSArrTime'] = fact_flights['CRSArrTime'].astype('int16')
fact_flights['CRSElapsedTime'] = fact_flights['CRSElapsedTime'].astype('int16')
fact_flights['DepDelay'] = fact_flights['DepDelay'].astype('int16')
fact_flights['ArrDelay'] = fact_flights['ArrDelay'].astype('int16')
fact_flights['Distance'] = fact_flights['Distance'].astype('int16')
fact_flights['TaxiIn'] = fact_flights['TaxiIn'].astype('int16')
fact_flights['TaxiOut'] = fact_flights['TaxiOut'].astype('int16')
fact_flights['ActualElapsedTime'] = fact_flights['ActualElapsedTime'].astype('int16')
fact_flights['AirTime'] = fact_flights['AirTime'].astype('int16')
fact_flights['Cancelled'] = fact_flights['Cancelled'].astype('int8')
fact_flights['Diverted'] = fact_flights['Diverted'].astype('int8')
fact_flights['CarrierDelay'] = fact_flights['CarrierDelay'].astype('int16')
fact_flights['WeatherDelay'] = fact_flights['WeatherDelay'].astype('int16')
fact_flights['NASDelay'] = fact_flights['NASDelay'].astype('int16')
fact_flights['SecurityDelay'] = fact_flights['SecurityDelay'].astype('int16')
fact_flights['LateAircraftDelay'] = fact_flights['LateAircraftDelay'].astype('int16')
fact_flights['CancellationCode'] = fact_flights['CancellationCode'].astype(str)

fact_flights.to_csv('fact_flights.csv', index=False)

print(fact_flights.columns.tolist())
print(fact_flights.shape)
```
